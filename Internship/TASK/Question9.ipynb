```python
data.head()
```

```python
data_1=data.select_dtypes('float64')
```

```python
a=pd.DataFrame(data.groupby('Country Name')['Sum'].sum())
```

```python
a
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent = 'app')
```

```python
a.columns
```

```python
a=pd.DataFrame.reset_index(a) 
```

```python
a
```

```python
lat_lon = []
for location in a['Country Name']:
    location = geolocator.geocode(location)
    if location is None:
        lat_lon.append(np.nan)
    else:
        geo = (location.latitude, location.longitude)
        lat_lon.append(geo)
```

```python
a['Cordinates']=lat_lon
```

```python
a.isnull().sum()
```

```python
a.dropna(inplace=True)
```

```python
a.head()
```

```python
latitude=[]
for i in a['Cordinates']:
    latitude.append(i[0])
```

```python
longtitude=[]
for i in a['Cordinates']:
    longtitude.append(i[1])
```

```python
a['Latitude']=latitude
a['Longitude']=longtitude
```
