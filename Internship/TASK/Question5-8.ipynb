# 6. AVERAGE AMOUNT OF DEBT ACROSS INDICATORS

```python
data.drop(['2022', '2023', '2024', '2025','2026', '2027', '2028'],inplace=True,axis=1)
```

```python
data.columns
```

```python
data_2 = data.select_dtypes('float64')
```

```python
data_2.head()
```

```python
data['Sum'] = data_2.sum(axis = 1)
```

```python
data.head()
```

```python
d = pd.DataFrame(data.groupby('Series Code')['Sum'].sum())
d.head(30)
```

### ABOVE DATAFRAME SHOWS AVERAGE DEBT ACROSS INDICATORS
